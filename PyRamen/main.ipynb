```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path
import csv
# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path(r'C:\Users\Office\Desktop\PyRamen_Resources_menu_data.csv')
sales_filepath = Path(r'C:\Users\Office\Desktop\PyRamen_Resources_sales_data.csv')

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []

# @TODO: Read in the menu data into the menu list
with open(menu_filepath, 'r') as file:
    menu_reader = csv.reader(file, delimiter = ',')
    manu_header = next(menu_reader)
    
    for items in menu_reader:
        menu.append(items)
    
    

# @TODO: Read in the sales data into the sales list
with open(sales_filepath, 'r') as file:
    sales_reader = csv.reader(file, delimiter = ',')
    sales_header = next(sales_reader)
    
    for items in sales_reader:
        sales.append(items)



# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}
for items in sales:
    sales_item = items[4]
    sales_quantity = items[3]
    
    if not sales_item in report:
        report[sales_item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0}
    else:
        continue



# Initialize a row counter variable
row_count = 0

# @TODO: Loop over every row in the sales list object

for items in sales:
    sales_item = items[4]
    sales_quantity = items[3]
    if sales_item in report:
        report[sales_item]['01-count'] += int(sales_quantity) 


    # Line_Item_ID,Date,Credit_Card_Number,Quantity,Menu_Item
    # @TODO: Initialize sales data variables
for items in sales:
    sales_item = items[4]
    sales_quantity = items[3]
    
    for items in menu:
        menu_item = items[0]
        menu_revenue = items[3]
        menu_cost = items[4]

# Aggregating Revenue and COGs        
        if sales_item == menu_item:
            report[menu_item]['02-revenue'] += int(menu_revenue) * int(sales_quantity)
        if sales_item == menu_item:
            report[menu_item]['03-cogs'] += int(menu_cost) * int(sales_quantity)

# Calculating Profit in the Dictionary            
for key, value in report.items():
    for items in value:
        revenue = report[key]['02-revenue']
        cogs = report[key]['03-cogs']
        
        if items == '04-profit':
            report[key][items] = revenue - cogs
        else:
            continue
    

# Creating Ouput file in csv    

output_path = (r'C:\Users\Office\Desktop\Pyramen_Assignment.csv')
with open(output_path, 'w') as file:
    file.write("PyRamen Assignment.\n")
    file.write("-----------------.\n")
    for key, value in report.items():
        file.write(f"{key} {value} \n")

        
print(report)
```
